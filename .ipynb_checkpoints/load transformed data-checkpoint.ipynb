```python
# import modules
import sqlite3
```

```python
laundry_mat_data = [
    {"product": "Detergent", "dollar_price_per_unit": 4.5, "quantity": 100, "total_cost": 200},
    {"product": "Dryer Sheets Box", "dollar_price_per_unit": 3.5, "quantity": 100, "total_cost": 350},
    {"product": "Washing Machine", "dollar_price_per_unit": 400, "quantity": 25, "total_cost": 10000},
    {"product": "Dryer", "dollar_price_per_unit": 400, "quantity": 25, "total_cost": 10000},
    {"product": "Laundry basket", "dollar_price_per_unit": 30, "quantity": 50, "total_cost": 1500},
    {"product": "Hanger", "dollar_price_per_unit": 1, "quantity": 200, "total_cost": 200}
]
```

```python
sqlite_db = "laundry_mat.sqlite"
```

```python
# Connect to the database
conn = sqlite3.connect(f'data/{sqlite_db}')
cursor = conn.cursor()
```

```python
# Truncate the existing data
cursor.execute("DELETE FROM laundry_mat;")
```

```python
cursor.execute("DROP TABLE laundry_mat;")
```

```python
cursor.execute("CREATE TABLE laundry_mat (product VARCHAR(255), dollar_price_per_unit DECIMAL, quantity INT, total_cost INT);")
```

```python
for row in laundry_mat_data:
    cursor.execute("INSERT INTO laundry_mat (product, dollar_price_per_unit, quantity, total_cost) VALUES (?, ?, ?, ?)",
    (row["product"], row["dollar_price_per_unit"], row["quantity"], row["total_cost"]))
```

```python
for row in laundry_mat_data:
    cursor.execute("INSERT OR REPLACE INTO laundry_mat (product, dollar_price_per_unit, quantity, total_cost) VALUES (?, ?, ?, ?)",
    (row["product"], row["dollar_price_per_unit"], row["quantity"], row["total_cost"]))
```

```python
conn.commit()
```

```python
cursor.execute("SELECT * FROM laundry_mat")
```

```python
# Fetch and print data
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
conn.close()
```
