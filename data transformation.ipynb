```python
def get_transformed_data(crash_data, vehicle_crash_data):

    # import data
    df_crashes = pd.read_csv(f"data/{crash_data}")
    df_vehicles = pd.read_csv(f"data/{vehicle_crash_data}")

    # drop columns where ALL values are null
    # modifies the dataframe directly (inplace=True)
    df_crashes.dropna(axis='columns', how='all', inplace=True)
    # drop rows with <2 non-null values
    df_crashes.dropna(axis='index', thresh=2, inplace=True)

    # fill missing values in the report_type column
    df_crashes.fillna(value={'report_type': 'ON SCENE'}, inplace=True)

    # merge crashes and vehicles data
    df = df_crashes.merge(right=df_vehicles, how='left', on='crash_record_id', suffixes=('_left', '_right'))
    df_agg = df.groupby('vehicle_type').agg({'crash_record_id': 'count'}).reset_index()

    # transform column names for output data
    vehicle_mapping = {'vehicle_type': 'vehicletypes'}
    df_agg = df_agg.rename(columns=vehicle_mapping)

    return df_crashes.info()
```

```python
# Read data from the source
def read_data_sources(source_name):
    df = pd.read_csv(f"data/{source_name}")
    return df
```

```python
# Drop columns where ALL values are null
def drop_null_columns(df):
    df.dropna(axis='columns', how='all', inplace=True)
    return df
```

```python
# Drop rows with <2 non-null values
def drop_null_rows(df):
    df.dropna(axis='index', thresh=2, inplace=True)
    return df
```

```python
# Fill missing values in the report_type column
def fill_missing_report_type(df):
    df.fillna(value={'report_type': 'ON SCENE'}, inplace=True)
    return df
```

```python
# Merge crashes and vehicles data
def merge_data(df_crashes, df_vehicles):
    df = df_crashes.merge(right=df_vehicles, how='left', on='crash_record_id', suffixes=('_left', '_right'))
    return df
```

```python
# Rename columns for output data
def rename_columns(df):
    vehicle_mapping = {'vehicle_type': 'vehicletypes'}
    df = df.rename(columns=vehicle_mapping)
    return df
```

```python
def read_data_pipeline(crash_data, vehicle_crash_data):
    try:
        df_crash = read_data_sources(crash_data)
        df_vehicle = read_data_sources(vehicle_crash_data)
    except Exception as e:
        logging.info("Exception in reading data pipeline")
    
    return df_crash, df_vehicle
```

```python
def drop_null_cols_pipeline(df_crash, df_vehicle):
    try:
        drop_null_columns(df_crash)
        drop_null_columns(df_vehicle)
    except Exception as e:
        logging.info("Exception in dropping null columns pipeline")
        
    return df_crash, df_vehicle
```

```python
def drop_null_rows_pipeline(df_crash, df_vehicle):
    try:
        drop_null_rows(df_crash)
        drop_null_rows(df_vehicle)
    except Exception as e:
        logging.info("Exception in dropping null rows pipeline")
        
    return df_crash, df_vehicle
```

```python
def fill_missing_values_pipeline(df_crash, df_vehicle):
    try:
        df_crash = fill_missing_report_type(df_crash)
        df_vehicle = fill_missing_report_type(df_vehicle)
    except Exception as e:
        logging.info("Exception in filling missing report type pipeline")
        
    return df_crash, df_vehicle
```

```python
def merge_data_pipeline(df_crash, df_vehicle):
    try:
        df_merged = merge_data(df_crash, df_vehicle)
    except Exception as e:
        logging.info("Exception in merge data pipeline")
        
    return df_merged
```

```python
def format_data_pipeline(df_merged):
    try:
        df_output = rename_columns(df_merged)
    except Exception as e:
        logging.info("Exception in renaming data column pipeline")
        
    return df_output
```
